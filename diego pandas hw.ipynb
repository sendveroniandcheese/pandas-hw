## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
item_count = purchase_data['Item ID'].nunique()
item_count

average_itemprice = purchase_data['Price'].mean()
average_itemprice

total_purchases = purchase_data['Purchase ID'].count()
total_purchases

total_revenue = purchase_data['Price'].sum()
total_revenue

purchasing_analysis = pd.DataFrame({"Item Count": [item_count], 
                                   "Average Price": [average_itemprice],
                                    "Total Purchases": [total_purchases],
                                   "Total Revenue": [total_revenue]
                                                                       })
purchasing_analysis
```

```python
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${0:,.0f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${0:,.0f}".format)

purchasing_analysis
```
